```python
import os
import shutil
import json
import ast
```

```python
for i in os.listdir('../code_forces/'):
    if (i != '.DS_Store'):
        if len(os.listdir('../code_forces/' + i)) < 40:
            print(i, len(os.listdir('../code_forces/' + i)))
```

```python
#Remove empty folder
walk = list(os.walk('../code_forces/'))
for path, _ ,_ in walk[::-1]:
    if (len(os.listdir(path))) == 0: os.rmdir(path)
```

```python
#Remove empty file in folder
for folder in os.listdir('../code_forces/'):
    if (folder != '.DS_Store'):
        for file in os.listdir('../code_forces/' + folder):
            if (file != '.DS_Store'):
                if not os.path.getsize('../code_forces/' + folder + '/' + file):
                    os.remove('../code_forces/' + folder + '/' + file)
```

```python
#Print Average number of files in folder
sum = 0
for folder in os.listdir('../code_forces/'):
    if (folder != '.DS_Store'):
        sum+=len(os.listdir('../code_forces/' + folder))
avg = sum / len(os.listdir('../code_forces/'))
print(avg)
print(sum)
```

```python
tags = dict()
with open('../../_tags.txt', 'r') as tag_file:
    content = tag_file.readlines()
    for i in content:
        i = ast.literal_eval(i)
        for problem in i:
           tags[problem] = i[problem] 
for problem in os.listdir('../../code_forces_6_3/'):
    with open('../../tags.txt','a') as t_file:
        try:
            if (tags[problem] != []):
                t_file.write('{\'' + problem + '\' : ' + str(tags[problem]) + '}')
                t_file.write('\n')
        except: print(problem)
```
